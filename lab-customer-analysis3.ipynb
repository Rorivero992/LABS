```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("marketing_customer_analysis.csv") 
```

```python
#copia de respaldo.
df2=df1.copy()
```

```python
df1.info()
```

```python
df1.describe()
```

```python
df1['Response'].hist(bins=10)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Sales Channel', hue='Response', data=df1)
plt.title('Response Rate by Sales Channel')
plt.show()
```

```python
sns.barplot(x="RESPONSE", y="TOTAL_CLAIM_AMOUNT", data=df1)
```

```python
#grafico de prueba
plt.figure(figsize=(8,6))
sns.countplot(x='Total Claim Amount', hue='Response', data=df1)
plt.title('Response Rate Total Claim Amount')
plt.show()
```

```python
sns.boxplot(x='Response', y='Total Claim Amount', data=df1)
plt.title('Response Rate by Total Claim Amount')
plt.show()
```

```python
#grafico de prueba
plt.figure(figsize=(8,6))
sns.countplot(x='Income', hue='Response', data=df1)
plt.title('Response Rate by Income')
plt.show()
```

```python
#grafico de prueba 
plt.figure(figsize=(10,6))
sns.countplot(x='INCOME', hue='RESPONSE', data=df1, width=3)
plt.title('Response Rate by Income')
plt.show()
```
