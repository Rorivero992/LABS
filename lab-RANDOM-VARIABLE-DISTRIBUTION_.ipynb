```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("we_fn_use_c_marketing_customer_value_analysis.csv")
```

```python
df_numerical=df.select_dtypes(include=np.number)
df_numerical
```

```python
for col in df_numerical.columns:
    sns.displot(df_numerical[col])
    plt.title(col)
    plt.show()
```

```python
def log_transfom_clean1(x):
    x = np.log(x)
    if np.isfinite(x):
        return x
    else:
        return 0
        
Customer_Lifetime_Value_log = list(map(log_transfom_clean1, df_numerical['Customer Lifetime Value']))
sns.distplot(Customer_Lifetime_Value_log)
plt.show()
```
